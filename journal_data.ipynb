```python
final = open("refined.txt", "r")
finalread = final.readlines()
```

```python
TI = []
AU = []
SO = []
PY = []
AB = []
au_t = 1
ti_t = 1
so_t = 1
py_t = 1
ab_t = 1
for line in finalread:
    split_lines = line.split(" ")
    if len(split_lines) > 0:
        # Appending Author information
        if split_lines[0] == "AU":
            list_authors = line[3:].split("  ")
            authors = ''
            for j in list_authors:
                author = j.rstrip("\n") + ";"
                authors = authors + author
            authors.rstrip(';')
            AU.append(authors)
            au_t = 0
        
        # Appending title information
        if split_lines[0] == "TI":
            title = ''
            title = title + line[3:].rstrip("\n")
            TI.append(title)
            ti_t = 0
        
        # Appending journal information
        if split_lines[0] == "SO":
            so = line[3:].rstrip("\n")
            SO.append(so)
            so_t = 0
            
        # Appending year information
        if split_lines[0] == "PY":
            py = line[3:].rstrip("\n")
            PY.append(py)
            py_t = 0
            
        # Appending abstract information
        if split_lines[0] == "AB":
            ab = line[3:].rstrip("\n")
            AB.append(ab)
            ab_t = 0
        
        # Checking information
        if split_lines[0].rstrip("\n") == "ER":
            if au_t==1 or ab_t==1 or py_t==1 or ti_t==1 or so_t==1:
                if au_t == 1:
                    AU.append(" ")
                    print("Missing AU")
                if ab_t == 1:
                    AB.append(" ")
                    print("Missing AB")
                if py_t == 1:
                    PY.append(" ")
                    print("Missing PY")
                if ti_t == 1:
                    TI.append(" ")
                    print("Missing TI")
                if so_t == 1:
                    SO.append(" ")
                    print("Missing SO")
                    print(TI[-1])
            else:
                au_t = 1
                ab_t = 1
                py_t = 1
                ti_t = 1
                so_t = 1
            
```

```python
data = [PY, SO, AU, TI, AB]

df = DataFrame(data).transpose()
df.columns = ['year','journal','author(s)', 'title', 'abstract']
```
